# Reverse Polish Notation (RPN) and Stack Visualizations

One of the most memorable parts of my AP CSA preparation was building and teaching concepts related to **stacks, queues, and Reverse Polish Notation (RPN)**.

Instead of simply studying definitions, I created visualizations that showed how stacks and queues behave in real time.

## Why This Helped

Visualizing the structures made concepts easier to understand:

* pushing onto a stack → adds to the top
* popping → removes the most recent value
* queues → process elements in order

This changed my understanding from:

> memorizing definitions → understanding behavior and logic

---

# Reverse Polish Notation (RPN)

Instead of writing:

```text id="x8b20w"
3 + 4
```

RPN writes expressions as:

```text id="6j4t8m"
3 4 +
```

The algorithm works entirely using a stack.

## Core Process

1. Read tokens left → right
2. Push numbers onto the stack
3. Pop values when encountering operators
4. Push the result back onto the stack

## What I Learned

Building the visualization helped me understand:

* stack-based computation
* why order matters for subtraction/division
* how data structures solve real problems logically

This was one of the first times computer science felt more like logical system design instead of memorization.

## FRQ 2017 Q1: Digits

https://shriya1401.github.io/shriyap_2025/2025/11/05/2026v1_IPYNB_2_.html

This FRQ focused on storing digits in an `ArrayList` and checking whether they formed a strictly increasing sequence.

### Important Lessons

* handling edge cases like `0`
* preserving digit order
* loop bounds (`size() - 1`)
* avoiding `IndexOutOfBoundsException`

```java id="7p57iw"
public boolean isStrictlyIncreasing() {
    for (int i = 0; i < digitList.size() - 1; i++) {
        if (digitList.get(i) >= digitList.get(i + 1)) {
            return false;
        }
    }
    return true;
}
```

### Key Takeaway

Small details and edge cases are often where most mistakes happen.

---

# Patterns I Noticed Across FRQs and MCQs

As I practiced more FRQs and MCQs, recurring patterns started appearing repeatedly.

| Pattern             | Example                    |
| ------------------- | -------------------------- |
| Stack usage         | RPN calculator             |
| Edge cases          | empty lists, single values |
| Loop traversal      | arrays and ArrayLists      |
| Helper method reuse | FRQ helper methods         |
| Condition logic     | SignedTextClass            |
| Off-by-one errors   | loop bounds                |
| Careful tracing     | MCQ runtime analysis       |

Instead of memorizing solutions, I started recognizing reusable structures and strategies.

---

# Growth Over Time

Looking back at my earliest FRQs compared to the later ones, the biggest improvement was not speed — it was confidence in reasoning.

Early on, I would immediately start coding and hope the logic worked.

Later, I learned to:

* trace examples first
* think about edge cases early
* slow down during MCQs
* recognize common patterns
* use helper methods intentionally

The problems became much less intimidating because I developed a structured process for approaching them.

---

# Final Reflection

Preparing for the AP CSA exam became much more than preparing for a test.

It became an experience centered around:

* teaching concepts
* visualizing systems
* recognizing patterns
* understanding logic deeply

The most meaningful part was not simply solving problems.

It was creating systems where:

> other people could learn from what I built.
