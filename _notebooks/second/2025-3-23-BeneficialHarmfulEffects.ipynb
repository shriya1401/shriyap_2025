## Classwork Questions

#### 1. Explain how the same computing innovation can have both beneficial and harmful effects depending on the context and user perspective.

**Example:** Facial recognition software can help find missing people and unlock phones, which is beneficial. However, it can also be used for mass surveillance and has been shown to have racial bias in some systems. One person might feel safer with it, while another might feel constantly watched or unfairly targeted. This shows how the same innovation can affect people in different ways depending on the context and perspective.

---

#### 2. Identify an example of a computing innovation that had unintended consequences. Explain what the intended use was and how the unintended effects emerged.

**Example:** Social media was originally designed to help people connect and share experiences. However, an unintended consequence is that it can spread misinformation quickly and create echo chambers. These effects emerged as platforms prioritized engagement and reach, often without considering how the algorithms might be manipulated or misused.

---

#### 3. How can programmers anticipate and mitigate the harmful effects of computing innovations? List at least three strategies based on the principles of responsible innovation.

1. **Test with diverse users** to make sure the technology works fairly for everyone.  
2. **Anticipate misuse scenarios** by thinking about how the innovation could be abused.  
3. **Build in safeguards** like privacy settings, content moderation, or opt-in features to protect users.

### Free Response

#### For the computing innovation of autonomous vehicles, analyze both beneficial and harmful effects across different contexts.

Autonomous vehicles can reduce traffic accidents, offer mobility for people who can’t drive, and make transportation more efficient. However, they also raise concerns about job loss in transportation industries, system failures, and ethical decision-making in crash scenarios. There are also privacy risks from data collected during rides. While helpful in safety and access, they can create complex new risks for society.

### Group Activity: Innovation Design Challenge

#### Identify a problem that could be addressed with computing:
**Cyberbullying** is a serious problem in schools and online spaces, and many students are afraid to speak up or ask for help.

#### Design a computing innovation solution:
We created an AI-powered app called **SafeSpace**, which detects harmful or bullying messages using machine learning. It alerts the user privately and provides links to support resources or allows anonymous reporting.

#### Analyze beneficial and harmful effects:
**Beneficial Effects:**  
- Helps identify cyberbullying early  
- Encourages students to seek help  
- Reduces long-term emotional harm  

**Harmful Effects:**  
- Might falsely flag harmless jokes  
- Could invade user privacy if data isn’t protected  
- May make users feel monitored

### Propose safeguards to minimize harmful effects:
- Use encrypted, anonymous detection to protect identities  
- Let users review flagged messages and adjust sensitivity  
- Allow human moderators to verify reports before taking action

### Assignment: Computing Innovation Impact Analysis

#### Choose a computing innovation not discussed in class:
**Innovation:** Augmented Reality (AR) in education.

#### Identify and describe the innovation:
Augmented reality blends digital content with the real world, like using AR glasses or apps to explore 3D models during class. It allows students to visualize complex subjects like anatomy, architecture, or chemistry in interactive ways.

#### Analyze beneficial effects across multiple categories:
- **Education:** Makes learning more interactive and memorable  
- **Creativity:** Encourages creative problem solving and design  
- **Accessibility:** Helps visual learners and students with learning differences

#### Analyze harmful effects across multiple categories:
- **Cost:** AR tools can be expensive and widen the digital divide  
- **Privacy:** Some AR apps track location and usage data  
- **Health:** Overuse may lead to eye strain or reduced attention spans

#### Explain an effect that could be viewed as both beneficial and harmful:
**Increased engagement**—while AR can make lessons more fun, it could also distract from deeper learning or make students dependent on visual effects.

#### Identify at least one unintended use or consequence:
AR filters meant for education could be misused on social media for unrealistic image enhancement, leading to self-esteem issues.

#### Propose measures to maximize benefits and minimize harms:
- Offer AR devices at school to ensure equal access  
- Include privacy settings and clear data usage terms  
- Limit non-educational content to keep students focused

### Why This Matters

Understanding both the beneficial and harmful effects of computing innovations helps us become better designers, programmers, and users. It prepares us to think critically, develop more ethical solutions, and anticipate the real-world impact of the tools we create.
