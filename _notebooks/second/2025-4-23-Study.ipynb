4-Week AP CSP Study Plan


# Week 1: Internet, Data, & Digital Information

*Focus Areas:*

- The Internet: IP, DNS, HTTP, TCP, packets, redundancy

- Digital data: binary, hex, file sizes, abstraction

- Data compression: lossy vs lossless

*Tasks:*

- Review Units 1 & 2 (College Board Framework)

- Watch: CrashCourse Computer Science – Internet

- Practice on Code.org (Lessons: 1–9)

- Quiz: Data conversions & compression examples

# Week 2: Programming, Algorithms, & Abstractions
*Focus Areas:*

- Variables, input/output, conditionals, loops

- Functions with parameters, return values

- Algorithms: searching, sorting, efficiency

*Tasks:*

- Review Unit 3 (Programming Fundamentals)

- Practice coding: Replit or CodeHS

- Build: A simple quiz app or calculator

- Learn AP pseudocode syntax (essential for the exam!)

# Week 3: Data, Cybersecurity, & Global Impacts
*Focus Areas:*

Big data & pattern recognition

Cybersecurity basics: encryption, phishing, TLS

Impacts of computing: innovations, bias, digital divide

*Tasks:*

- Review Units 4 & 5

- Research 1 real-world innovation for written prompts

- Watch: AP CSP Cybersecurity YouTube series

- Create flashcards for tech terms (e.g. certificate authority, symmetric encryption)

# Week 4:  Full Practice

*Focus Areas:*

- Practice MCQs + FRQs

- Simulate test conditions

*Tasks:*

- Go over sample Create Tasks + scoring

- Take 1 full-length practice exam (AP Classroom or 5 Steps to a 5)

- Review missed questions and redo them

- Quiz yourself on pseudocode logic and data reasoning
