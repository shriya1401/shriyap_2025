<h2 style="color:#38bdf8; border-bottom:2px solid #38bdf8; padding-bottom:6px;">
Biggest Takeaway
</h2>

<blockquote style="border-left:4px solid #38bdf8; padding-left:16px; color:#f9fafb; font-size:20px; font-weight:bold;">
People don’t buy products. They buy personality and emotion.
</blockquote>

<p>
The biggest lesson I learned from Elijah Johnson was that technical skill alone is not enough to create success. People connect with the story, confidence, personality, and emotion behind a product. Even if someone creates amazing technology, they still need communication, branding, and relationships to make people believe in the idea.
</p>

<h2 style="color:#38bdf8; border-bottom:2px solid #38bdf8; padding-bottom:6px;">
Key Takeaways
</h2>

<h3 style="color:#ffffff;">1. Start Before You Are Ready</h3>
<p>
Elijah emphasized that successful people do not wait for perfect conditions before taking action. Instead, they begin building early and improve while learning. This connects to computer science because real growth happens through actually creating projects, not only planning them.
</p>

<h3 style="color:#ffffff;">2. Consistency Matters More Than Motivation</h3>
<p>
A major takeaway was that discipline creates long-term success. Motivation changes every day, but consistency builds real progress. Elijah’s example of starting with something small, like two pushups a day, showed that success begins with simply showing up.
</p>

<h3 style="color:#ffffff;">3. Relationships Create Opportunities</h3>
<p>
Elijah explained that simple relationships can take people further than they expect. His connections helped SafeBallot grow, including partnerships involving cybersecurity and technical support. This showed me that networking can be just as important as technical ability.
</p>

<h3 style="color:#ffffff;">4. Rejection Is Part of Success</h3>
<blockquote style="border-left:4px solid #22c55e; padding-left:16px; color:#dcfce7; font-weight:bold;">
It’s not about the 99 no’s. It’s about the one yes.
</blockquote>

<p>
His solar sales experience showed that rejection is part of the process. Whether someone is applying for opportunities, presenting an idea, or building a business, one rejection should not stop them from continuing forward.
</p>

<h3 style="color:#ffffff;">5. Build While You Learn</h3>
<p>
Another important takeaway was that learning becomes more valuable when it is applied in real life. Elijah encouraged students to take action, use resources strategically, and apply skills immediately instead of only taking notes or waiting for permission.
</p>

<h2 style="color:#38bdf8; border-bottom:2px solid #38bdf8; padding-bottom:6px;">
Connection to Computer Science
</h2>

<p>
This event connected strongly to computer science because Elijah discussed how frontend systems, backend systems, cybersecurity, and branding all work together to create successful technology. SafeBallot showed that technology is not only about writing code, but also about solving real problems and building trust with users.
</p>

<p>
One idea that stood out was the importance of making something better, faster, and cheaper. This connects to innovation in computer science because many successful products improve existing systems instead of inventing something completely new.
</p>

<h2 style="color:#38bdf8; border-bottom:2px solid #38bdf8; padding-bottom:6px;">
Personal Reflection
</h2>

<p>
This presentation changed the way I think about success. Before this event, I mainly thought about achievement through technical skills, but Elijah showed that communication, confidence, discipline, networking, and consistency are equally important.
</p>

<p>
The speaker also encouraged me to think more like a founder instead of just a student waiting for permission. I learned that you do not need to know every step of the future. You only need to know your starting point, your goal, and your next step.
</p>

<h2 style="color:#ffffff; margin-top:0;">Final Reflection</h2>

<p style="color:#d1d5db;">
Overall, this event taught me that success comes from taking action, building discipline, creating relationships, and applying knowledge in real-life situations.
</p>

<p style="color:#38bdf8; font-size:20px; font-weight:bold;">
Start before you are ready. Build while you learn.
