## FRQ 2017 Q1: Digits

[View Blog Post](https://shriya1401.github.io/shriyap_2025/2025/11/05/2026v1_IPYNB_2_.html)

This FRQ focused on storing digits inside an `ArrayList` and checking whether they formed a strictly increasing sequence.

### Important Lessons

* handling `0` correctly as an edge case
* preserving digit order
* avoiding `IndexOutOfBoundsException`
* using proper loop bounds (`size() - 1`)

```java
public boolean isStrictlyIncreasing() {
    for (int i = 0; i < digitList.size() - 1; i++) {
        if (digitList.get(i) >= digitList.get(i + 1)) {
            return false;
        }
    }
    return true;
}
```

### Key Takeaway

Small details and edge cases are where most FRQ mistakes happen.

# Project-Based Review: Stacks, Queues, and RPN

One of the most useful projects I built while studying was a visualization for stacks, queues, and Reverse Polish Notation (RPN).

Instead of simply memorizing definitions, I built interactive systems that showed how the structures worked step-by-step.

## What I Learned

* stacks use Last In, First Out
* queues use First In, First Out
* RPN uses stacks to evaluate expressions
* order matters for subtraction and division

Example:

```text
3 4 +
```

represents:

```text
3 + 4
```

Building these visualizations helped me understand how data structures solve real computational problems logically.

# Growth Over Time

Looking back at my earliest FRQs compared to later ones, the biggest improvement was not speed — it was confidence in reasoning.

At the beginning, I would immediately start coding and hope the logic worked.

Over time, I developed a much more structured process:

1. read the full question carefully
2. identify helper methods
3. trace examples manually
4. think about edge cases
5. then begin coding

This process made both MCQs and FRQs feel much less intimidating.
