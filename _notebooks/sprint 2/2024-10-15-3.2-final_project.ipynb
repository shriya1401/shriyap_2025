```python
import random

# Class to manage the game
class BasketballGame:
    def __init__(self):
        self.player_name = input("Enter your name: ")
        self.total_points = 0
        self.shots_taken = 0
        self.turns_left = 5
        self.shot_results = []
        self.streak = 0  # Track consecutive successful shots
        self.success_rate = 0

    # Simulate taking a shot
    def take_shot(self):
        if self.turns_left <= 0:
            print("Game Over! You've run out of turns.")
            self.end_game()
            return

        # Simulate shot success (70% chance)
        success_rate = random.randint(1, 10)
        shot_successful = success_rate > 3  # 70% chance to score

        self.shots_taken += 1
        self.shot_results.append(shot_successful)

        # Update total points and streak
        if shot_successful:
            self.total_points += 1
            self.streak += 1
            print(f"Great shot! You've scored a point. Current Streak: {self.streak}")
        else:
            self.streak = 0
            print("Missed the shot. Better luck next time.")

        self.turns_left -= 1
        self.calculate_success_rate()
        self.update_stats()

    # Update game stats after each shot
    def update_stats(self):
        stats = f"""
        {self.player_name}'s Stats:
        Total Points: {self.total_points}
        Shots Taken: {self.shots_taken}
        Turns Left: {self.turns_left}
        Success Rate: {self.success_rate:.2f}%
        Shot Results: {', '.join(map(lambda x: 'Hit' if x else 'Miss', self.shot_results))}
        """
        print(stats)

    # Calculate the success rate
    def calculate_success_rate(self):
        if self.shots_taken > 0:
            self.success_rate = (self.total_points / self.shots_taken) * 100

    # End the game and thank the player
    def end_game(self):
        print(f"Thanks for playing, {self.player_name}! You scored {self.total_points} points.")
        self.reset_game()

    # Reset the game
    def reset_game(self):
        self.total_points = 0
        self.shots_taken = 0
        self.turns_left = 5
        self.shot_results = []
        self.streak = 0
        self.success_rate = 0
        print("The game has been reset. You can start a new game!")

# Function to run the game
def run_game():
    game = BasketballGame()

    while game.turns_left > 0:
        action = input("Type 'shot' to take a shot, or 'exit' to quit the game: ").lower()

        if action == 'shot':
            game.take_shot()
        elif action == 'exit':
            game.end_game()
            break
        else:
            print("Invalid command. Please type 'shot' or 'exit'.")

    if game.turns_left == 0:
        game.end_game()

# Start the game
run_game()
```
